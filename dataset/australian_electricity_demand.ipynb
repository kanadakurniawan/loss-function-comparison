```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from scipy.stats import skew, kurtosis, shapiro, norm
from statsmodels.tsa.stattools import adfuller
import seaborn as sns

# Fungsi untuk membaca file TSF dari URL
def read_tsf_from_url(url):
    response = requests.get(url)
    lines = response.text.splitlines()

    series_data = []
    reading_data = False

    for line in lines:
        if line.strip() == "" or line.startswith("#"):
            continue

        if line.startswith("@data"):
            reading_data = True
            continue

        if reading_data:
            series_data.append(line.strip())

    return series_data

# Parsing time series data
def parse_series_data(series_data):
    parsed_data = []
    state_names = []
    for row in series_data:
        parts = row.split(":")
        if len(parts) >= 4:
            state_names.append(parts[1])  # Extract the state name
            time_series = list(map(float, parts[3].split(","))) if "," in parts[3] else []
            parsed_data.append(time_series)
    return state_names, parsed_data

# Uji stasioneritas menggunakan Augmented Dickey-Fuller Test
def check_stationarity(ts, state_name):
    result = adfuller(ts)
    print(f"\nADF Test for {state_name}:")
    print(f"ADF Statistic: {result[0]:.2f}")
    print(f"p-value: {result[1]:.4f}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"   {key}: {value:.2f}")
    if result[1] <= 0.05:
        print("The series is likely stationary.")
    else:
        print("The series is likely non-stationary.")

# Dekomposisi data time series
def decompose_time_series(ts, state_name):
    from statsmodels.tsa.seasonal import seasonal_decompose
    ts_series = pd.Series(ts).fillna(0)
    decomposition = seasonal_decompose(ts_series, model='additive', period=24*7)  # Assuming weekly seasonality

    plt.figure(figsize=(12, 8))
    plt.subplot(411)
    plt.plot(decomposition.observed, label='Observed')
    plt.legend(loc='upper left')
    plt.subplot(412)
    plt.plot(decomposition.trend, label='Trend')
    plt.legend(loc='upper left')
    plt.subplot(413)
    plt.plot(decomposition.seasonal, label='Seasonal')
    plt.legend(loc='upper left')
    plt.subplot(414)
    plt.plot(decomposition.resid, label='Residual')
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.suptitle(f"Decomposition for {state_name}", y=1.02)
    plt.show()

# Uji normalitas menggunakan Shapiro-Wilk Test
def check_normality(ts, state_name):
    stat, p = shapiro(ts)
    print(f"\nNormality Test for {state_name}:")
    print(f"Shapiro-Wilk Statistic: {stat:.4f}")
    print(f"p-value: {p:.4f}")
    if p > 0.05:
        print("The data likely follows a normal distribution.")
    else:
        print("The data does not follow a normal distribution.")

    # Plot histogram with fitted normal curve
    plt.figure(figsize=(10, 5))
    sns.histplot(ts, kde=False, stat="density", bins=30, label="Histogram", color="blue")
    mean, std_dev = np.mean(ts), np.std(ts)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mean, std_dev)
    plt.plot(x, p, 'k', linewidth=2, label="Normal Curve")
    plt.title(f"Histogram and Normal Curve for {state_name}")
    plt.legend()
    plt.grid()
    plt.show()

# URL raw file dari GitHub
url = "https://raw.githubusercontent.com/kanadakurniawan/loss-function-comparison/708bc046c7e760d74b7fccafb97a0cc203ff1f6e/database/australian_electricity_demand_dataset.tsf"

# Read the TSF file from the URL
raw_data = read_tsf_from_url(url)
states, time_series_data = parse_series_data(raw_data)

# Statistical analysis with handling for empty series
series_lengths = [len(ts) for ts in time_series_data]
means = [np.mean(ts) if len(ts) > 0 else np.nan for ts in time_series_data]
medians = [np.median(ts) if len(ts) > 0 else np.nan for ts in time_series_data]
std_devs = [np.std(ts) if len(ts) > 0 else np.nan for ts in time_series_data]
mins = [np.min(ts) if len(ts) > 0 else np.nan for ts in time_series_data]
maxs = [np.max(ts) if len(ts) > 0 else np.nan for ts in time_series_data]
skews = [skew(ts) if len(ts) > 0 else np.nan for ts in time_series_data]
kurtoses = [kurtosis(ts) if len(ts) > 0 else np.nan for ts in time_series_data]

statistical_summary = pd.DataFrame({
    "State": states,
    "Series Length": series_lengths,
    "Mean": means,
    "Median": medians,
    "Std Dev": std_devs,
    "Min": mins,
    "Max": maxs,
    "Skewness": skews,
    "Kurtosis": kurtoses
})

print("\nStatistical Summary:")
print(statistical_summary)

# Detailed analysis of time series
for i, ts in enumerate(time_series_data):
    if len(ts) > 0:
        print(f"\nDetailed Analysis for {states[i]}:")
        print(f"Total Data Points: {len(ts)}")
        print(f"Mean: {np.mean(ts):.2f}, Median: {np.median(ts):.2f}, Std Dev: {np.std(ts):.2f}")
        print(f"Min: {np.min(ts):.2f}, Max: {np.max(ts):.2f}")
        print(f"First Value: {ts[0]:.2f}, Last Value: {ts[-1]:.2f}")
        print(f"Range: {np.ptp(ts):.2f}, Variance: {np.var(ts):.2f}")
        print(f"Skewness: {skew(ts):.2f}, Kurtosis: {kurtosis(ts):.2f}")

        # Plot time series
        plt.figure(figsize=(10, 5))
        plt.plot(ts, label=f"{states[i]}")
        plt.title(f"Time Series Analysis for {states[i]}")
        plt.xlabel("Time")
        plt.ylabel("Electricity Demand")
        plt.legend()
        plt.grid()
        plt.show()

        # Autocorrelation
        plt.figure(figsize=(10, 5))
        pd.plotting.autocorrelation_plot(pd.Series(ts).fillna(0))
        plt.title(f"Autocorrelation for {states[i]}")
        plt.show()

        # Moving Average
        window_size = 100
        moving_avg = pd.Series(ts).rolling(window=window_size).mean()
        plt.figure(figsize=(10, 5))
        plt.plot(ts, label="Original")
        plt.plot(moving_avg, label=f"{window_size}-point Moving Avg")
        plt.title(f"Moving Average for {states[i]}")
        plt.xlabel("Time")
        plt.ylabel("Electricity Demand")
        plt.legend()
        plt.grid()
        plt.show()

        # Stationarity Test
        check_stationarity(ts, states[i])

        # Decomposition
        decompose_time_series(ts, states[i])

        # Normality Test
        check_normality(ts, states[i])
    else:
        print(f"No valid data for {states[i]}.")

# Visualization of the first time series
if time_series_data and len(time_series_data[0]) > 0:
    plt.figure(figsize=(12, 6))
    plt.plot(time_series_data[0])
    plt.title(f"Visualization of Time Series for {states[0]}")
    plt.xlabel("Time")
    plt.ylabel("Electricity Demand")
    plt.show()
else:
    print("No valid time series data found.")
```
